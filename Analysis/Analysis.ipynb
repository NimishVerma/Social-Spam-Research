# 3. Top 10 spammers, their categories, number of posts

```python
import pymongo
client = pymongo.MongoClient()
db = getattr(client, 'clustered_data')
print(db.list_collection_names())
posts = db.posts
curs = posts.find()
user_dict = {}
```

```python


for i, post in enumerate(curs):
    if post['user_id'] == -1:
        continue
    
    if post['user_id'] not in user_dict:
        abc = set()
        abc.add(post['category'])
        user_dict[post['user_id']] = {"count": 1, "categories": abc}
    else:
        user_dict[post['user_id']]['count'] += 1
        user_dict[post['user_id']]['categories'].add(post['category'])
        
#     sys.stdout.write("\r%i" % i)
#     sys.stdout.flush()
```

```python
## user_dict

sorted(user_dict.items(), key=lambda x: x[1]['count'], reverse=True)
```

# 6. Top Hashtags 

```python
import pymongo
client = pymongo.MongoClient()
db = getattr(client, 'clustered_data')
print(db.list_collection_names())
posts = db.posts
curs = posts.find()
hash_dic = {}
```

```python

for i, post in enumerate(curs):
    try:
        for hashtag in post['hashtags']:
            if hashtag not in hash_dic:

                hash_dic[hashtag] = 1
            else:
                hash_dic[hashtag] += 1
    except:
        pass
        
        
```

```python


sorted(hash_dic.items(), key=operator.itemgetter(1), reverse=True)
```

# 7. Most Prominent Spam Category as per Platform

```python
import pymongo
client = pymongo.MongoClient()
db = getattr(client, 'clustered_data')
print(db.list_collection_names())
posts = db.posts
curs = posts.find()
spam_category_per_platform = {}
```

```python
for i, post in enumerate(curs):
    
    if post['source'] not in spam_category_per_platform:
        
        spam_category_per_platform[post['source']] = {post['category']:1}
    else:
        if post['category'] not in spam_category_per_platform[post['source']]:
            spam_category_per_platform[post['source']][post['category']] = 1
        else:
            spam_category_per_platform[post['source']][post['category']] += 1
            
    
#     sys.stdout.write("\r%i" % i)
#     sys.stdout.flush()
```

```python
spam_category_per_platform
```

```python
flickr_sum = 0
twitter_sum = 0
tumblr_sum = 0

for platform, dic in spam_category_per_platform.items():
    if platform == 'flickr':
        for cat, count in dic.items():
            flickr_sum += count
    elif platform == 'twitter':
        for cat, count in dic.items():
            twitter_sum += count
    elif platform == 'tumblr':
        for cat, count in dic.items():
            tumblr_sum += count
    
```

```python
print(flickr_sum, twitter_sum, tumblr_sum)
```

```python
for platform, dic in spam_category_per_platform.items():
    for cat, count in dic.items():
        if platform == 'flickr':
            
            spam_category_per_platform[platform][cat] = (count*100) / flickr_sum
        elif platform == 'twitter':

            spam_category_per_platform[platform][cat] = (count*100) / twitter_sum
        elif platform == 'tumblr':
            
            spam_category_per_platform[platform][cat] = (count*100) / tumblr_sum
        
        
```

# 8. Same phone number in posts from multiple platforms.

```python
import pymongo
client = pymongo.MongoClient()
db = getattr(client, 'clustered_data')
print(db.list_collection_names())
posts = db.posts
curs = posts.find()
flickr_s = set()
twitter_s = set()
tumblr_s = set()

for i, post in enumerate(curs):
    if post['source'] == 'flickr':
        for elem in post['numbers'].split(','):
            flickr_s.add(elem)
            
    if post['source'] == 'twitter':
        for elem in post['numbers'].split(','):
            twitter_s.add(elem)
            
    if post['source'] == 'tumblr':
        for elem in post['numbers'].split(','):
            tumblr_s.add(elem)
    
```

```python
len(list(tumblr_s.intersection(twitter_s)))
```

# 9. Maximum number of people related to same campaign - same phone number - different ID

```python
import pymongo
client = pymongo.MongoClient()
db = getattr(client, 'clustered_data')
print(db.list_collection_names())
posts = db.posts
curs = posts.find()
phone_number_dict = {}

for i, post in enumerate(curs):
    for number in post['numbers'].split(','):
        if number == '' or number == '2018-2019' or number == '000\n2016':
            continue

        if number not in phone_number_dict:
            abc = set()
            abc.add(post['user_id'])
            phone_number_dict[number] = abc
        else:
            phone_number_dict[number].add(post['user_id'])
    
```

```python
campaign_size_dict = {}
for k, v in phone_number_dict.items():
#     print(k, v)
    campaign_size_dict[k] = len(list(v))
```

```python

sorted(campaign_size_dict.items(), key=operator.itemgetter(1), reverse=True)
```
