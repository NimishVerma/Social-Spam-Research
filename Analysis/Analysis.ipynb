# 3. Top 10 spammers, their categories, number of posts

```python
import pymongo
client = pymongo.MongoClient()
db = getattr(client, 'clustered_data')
print(db.list_collection_names())
posts = db.posts
curs = posts.find()
user_dict = {}
```

```python
for i, post in enumerate(curs):
    if post['user_id'] not in user_dict:
        abc = set()
        abc.add(post['category'])
        user_dict[post['user_id']] = {"count": 1, "categories": abc}
    else:
        user_dict[post['user_id']]['count'] += 1
        user_dict[post['user_id']]['categories'].add(post['category'])
```

```python
import operator

sorted(user_dict.items(), key=operator.itemgetter(1), reverse=True)
```
